```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
theta = np.random.uniform(0,np.pi,n)
phi = np.random.uniform(0,2*np.pi,n)

x = np.sin(theta)*np.cos(phi)
y = np.sin(theta)*np.sin(phi)
z = np.cos(theta)

# Creating figure 
fig = plt.figure(figsize = (10, 7)) 
ax = plt.axes(projection ="3d") 
  
# Creating plot 
ax.scatter3D(x, y, z, s=1)
# ax.view_init(elev=0., azim=0)
plt.title("Naive Sampling") 
  
# show plot 
plt.show() 
```

```python
n = 10000
t = np.random.uniform(-1,1,n)
theta = np.arccos(t)
phi = np.random.uniform(0,2*np.pi,n)

x = np.sin(theta)*np.cos(phi)
y = np.sin(theta)*np.sin(phi)
z = np.cos(theta)

# Creating figure 
fig = plt.figure(figsize = (10, 7)) 
ax = plt.axes(projection ="3d") 
  
# Creating plot 
ax.scatter3D(x, y, z, s=1)
# ax.view_init(elev=0., azim=0)
plt.title("Uniform Sampling") 
  
# show plot 
plt.show() 
```
